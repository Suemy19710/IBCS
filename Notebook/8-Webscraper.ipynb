```python
from serpapi import GoogleSearch
import requests, os
from PIL import Image
import imagehash

query = "corporate business graph IBSC style"

params = {
    "engine": "google",
    "q": query,
    "tbm": "isch",
    "ijn": "0",
    "api_key": "aba789da634d21ff41f1262a70b71557ccf2541ee4da75feb1e06382d652feb7"
}

def clean_image(path, min_size=200):
    try:
        img = Image.open(path).convert("RGB")

        if img.width < min_size or img.height < min_size:
            os.remove(path)
            return None

        img = img.resize((512, 512))
        img.save(path, "JPEG", quality=90)

        return imagehash.phash(img)

    except Exception:
        os.remove(path)
        return None

search = GoogleSearch(params)
results = search.get_dict()

save_path = "google_ibsc_images"
os.makedirs(save_path, exist_ok=True)

hashes = set()
count = 0

for image in results["images_results"]:
    url = image["original"]

    try:
        data = requests.get(url, timeout=5).content
        filename = os.path.join(save_path, f"{count}.jpg")

        with open(filename, "wb") as f:
            f.write(data)

        print("Downloaded:", filename)

        # Clean + dedupe
        file_hash = clean_image(filename)

        if file_hash is None:
            print("Removed (bad image):", filename)
            continue

        if file_hash in hashes:
            print("Duplicate removed:", filename)
            os.remove(filename)
        else:
            hashes.add(file_hash)
            print("Cleaned:", filename)
            count += 1

    except Exception as e:
        print("Failed:", url)
```
